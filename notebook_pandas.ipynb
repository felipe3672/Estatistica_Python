```python
import pandas as pd
import numpy as np
import timeit

# Criação de um DataFrame grande para teste
data = {'Suites': np.random.randint(0, 10, size=1000000)}
df = pd.DataFrame(data)

# Teste de desempenho para filtragem direta
def filter_direct():
    return df[df['Suites'] >= 1]

# Teste de desempenho para query
def filter_query():
    return df.query('Suites >= 1')

# Medindo o tempo de execução
time_direct = timeit.timeit(filter_direct, number=10)
time_query = timeit.timeit(filter_query, number=10)

print(f'Tempo de execução - Filtragem direta: {time_direct:.4f} segundos')
print(f'Tempo de execução - Query: {time_query:.4f} segundos')
```
